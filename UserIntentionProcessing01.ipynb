```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
#plt.rc('text', usetex = True)
```

```python
data = pd.read_csv('User_Intention_test01.csv')
```

```python
data.plot()
```

```python
sig1 = data['Signal-1'][50:500]
sig2 = data['Signal-2'][50:500]
sig3 = data['Signal-3'][50:500]
sig4 = data['Signal-4'][50:500]*100
plt.plot(sig1); plt.plot(sig2); plt.plot(sig3); plt.plot(sig4)
plt.grid();
plt.xlabel('Sample Index ($f_s$ = 30 Hz)')
plt.ylabel('Current [mA]')
```

```python
t1=70; t2=180
plt.plot(sig1[t1:t2])
plt.plot(sig2[t1:t2])
plt.plot(sig3[t1:t2])
plt.plot(sig4[t1:t2])
plt.axvline(170, 0.02, 0.98, color='gray', linestyle='--', linewidth=1)
plt.axvline(176, 0.1, 0.9, color='gray', linestyle='-.', linewidth=1)
#plt.grid()
plt.xlabel('Sample Index ($f_s$ = 30 Hz)')
plt.ylabel('Current [mA]')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
#plt.rc('text', usetex = True)
```

```python
data = pd.read_csv('Pressure_Drift_Test01.csv')
```

```python
data.plot(figsize = (10, 6), legend = False, xlabel = 'Sample Index ($f_s$ = 60 Hz)',
          ylabel = 'AD values [a.u.]', grid = True)
```

```python
sig7 = data['Signal-7'][0:1100]
plt.plot(sig7);
plt.grid();
plt.xlabel('Sample Index ($f_s$ = 60 Hz)')
plt.ylabel('AD values [a.u.]')
plt.axvline(380, 0.02, 0.98, color='gray', linestyle='--', linewidth=1)
plt.axvline(850, 0.02, 0.98, color='gray', linestyle='--', linewidth=1)
```

```python
sig7[380:850].mean(), sig7[380:850].std()
```

```python
sig1 = data['Signal-1'][4500:6000]
plt.plot(sig1);
plt.grid();
plt.xlabel('Sample Index ($f_s$ = 60 Hz)')
plt.ylabel('AD values [a.u.]')
plt.axvline(5020, 0.02, 0.98, color='gray', linestyle='--', linewidth=1)
plt.axvline(5490, 0.02, 0.98, color='gray', linestyle='--', linewidth=1)
```

```python
sig1[500:1000].mean(),sig1[500:1000].mean()*0.05, sig1[500:1000].std(), sig1[500:1000].std()/sig1[500:1000].mean()
```
